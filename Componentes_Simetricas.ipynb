# Componentes Simétricas — Transformación de Fortescue

Cálculo de componentes de secuencia (positiva, negativa, cero) a partir de valores de fase desbalanceados, y reconstrucción inversa (secuencia → fase) como verificación.

La demostración analítica de $A^{-1} = \frac{1}{3}A^*$ se encuentra en `demostracion.md`, en el mismo repositorio.

## 1. Definición del operador y matrices base

$a = 1\angle120^\circ$, y las matrices de transformación:

$$A = \begin{bmatrix}1&1&1\\1&a^2&a\\1&a&a^2\end{bmatrix}, \qquad A^{-1} = \frac{1}{3}A^* = \frac{1}{3}\begin{bmatrix}1&1&1\\1&a&a^2\\1&a^2&a\end{bmatrix}$$

```python
import numpy as np

def fasor(z):
    mag = np.abs(z)
    ang = np.degrees(np.angle(z))
    return f"{mag:.4f} \u2220 {ang:.2f}\u00b0"

a = np.exp(1j * np.deg2rad(120))

A = np.array([
    [1, 1,    1],
    [1, a**2, a],
    [1, a,    a**2]
])

A_conj = np.conj(A)
A_inv = (1/3) * A_conj

print("A =\n", np.round(A, 4))
print("\nA_inv =\n", np.round(A_inv, 4))
```

## 2. Fase → Secuencia

$$\begin{bmatrix}I_a^{(0)}\\I_a^{(1)}\\I_a^{(2)}\end{bmatrix} = A^{-1}\begin{bmatrix}I_a\\I_b\\I_c\end{bmatrix}$$

Datos de entrada (ajustar según el caso a resolver):

```python
# Entradas - CORRIENTES
Ia = 10 + 0j
Ib = 10 * np.exp(1j * np.deg2rad(180))
Ic = 0 + 0j

# Entradas - VOLTAJES
Va = 10 + 0j
Vb = 10 * np.exp(1j * np.deg2rad(180))
Vc = 0 + 0j

I_abc = np.array([Ia, Ib, Ic])
V_abc = np.array([Va, Vb, Vc])

I_012 = A_inv @ I_abc
V_012 = A_inv @ V_abc

print("I0 =", fasor(I_012[0]))
print("I1 =", fasor(I_012[1]))
print("I2 =", fasor(I_012[2]))
print()
print("V0 =", fasor(V_012[0]))
print("V1 =", fasor(V_012[1]))
print("V2 =", fasor(V_012[2]))
```

## 3. Secuencia → Fase (reconstrucción y verificación)

$$\begin{bmatrix}I_a\\I_b\\I_c\end{bmatrix} = A\begin{bmatrix}I_a^{(0)}\\I_a^{(1)}\\I_a^{(2)}\end{bmatrix}$$

Si la transformación es consistente, reconstruir desde las componentes de secuencia debe devolver exactamente los valores originales.

```python
I_abc_reconstruido = A @ I_012
V_abc_reconstruido = A @ V_012

print("I_abc reconstruido:", np.round(I_abc_reconstruido, 4))
print("¿Coincide con original?", np.allclose(I_abc, I_abc_reconstruido))
print()
print("V_abc reconstruido:", np.round(V_abc_reconstruido, 4))
print("¿Coincide con original?", np.allclose(V_abc, V_abc_reconstruido))
```

## 4. Componentes de fase b y c por secuencia

Descomposición de cada fase en sus tres componentes de secuencia individuales, aplicando la rotación correspondiente ($a$, $a^2$) a cada una.

```python
# Corrientes por secuencia y fase
Ia_0, Ia_1, Ia_2 = I_012[0], I_012[1], I_012[2]
Ib_0, Ib_1, Ib_2 = I_012[0], a**2 * I_012[1], a * I_012[2]
Ic_0, Ic_1, Ic_2 = I_012[0], a * I_012[1], a**2 * I_012[2]

# Voltajes por secuencia y fase
Va_0, Va_1, Va_2 = V_012[0], V_012[1], V_012[2]
Vb_0, Vb_1, Vb_2 = V_012[0], a**2 * V_012[1], a * V_012[2]
Vc_0, Vc_1, Vc_2 = V_012[0], a * V_012[1], a**2 * V_012[2]

print("--- CORRIENTES DE SECUENCIA POSITIVA ---")
print("Ia(1) =", fasor(Ia_1))
print("Ib(1) =", fasor(Ib_1))
print("Ic(1) =", fasor(Ic_1))
print()
print("--- CORRIENTES DE SECUENCIA NEGATIVA ---")
print("Ia(2) =", fasor(Ia_2))
print("Ib(2) =", fasor(Ib_2))
print("Ic(2) =", fasor(Ic_2))
print()
print("--- CORRIENTES DE SECUENCIA CERO ---")
print("Ia(0) =", fasor(Ia_0))
print("Ib(0) =", fasor(Ib_0))
print("Ic(0) =", fasor(Ic_0))
print()
print("--- VOLTAJES DE SECUENCIA POSITIVA ---")
print("Va(1) =", fasor(Va_1))
print("Vb(1) =", fasor(Vb_1))
print("Vc(1) =", fasor(Vc_1))
print()
print("--- VOLTAJES DE SECUENCIA NEGATIVA ---")
print("Va(2) =", fasor(Va_2))
print("Vb(2) =", fasor(Vb_2))
print("Vc(2) =", fasor(Vc_2))
print()
print("--- VOLTAJES DE SECUENCIA CERO ---")
print("Va(0) =", fasor(Va_0))
print("Vb(0) =", fasor(Vb_0))
print("Vc(0) =", fasor(Vc_0))
```
